```
# åˆ›å»ºè®­ç»ƒæ•°æ®
# ç‰¹å¾ï¼šé‡é‡ï¼ˆå…‹ï¼‰å’Œé¢œè‰²ï¼ˆ0è¡¨ç¤ºçº¢è‰²ï¼Œ1è¡¨ç¤ºæ©™è‰²ï¼‰
X_train = np.array([
    [150, 0],  # çº¢è‰²è‹¹æœ
    [200, 0],  # çº¢è‰²è‹¹æœ
    [180, 0],  # çº¢è‰²è‹¹æœ
    [300, 1],  # æ©™è‰²æ©™å­
    [250, 1],  # æ©™è‰²æ©™å­
    [280, 1],  # æ©™è‰²æ©™å­
])

# æ ‡ç­¾ï¼š0è¡¨ç¤ºè‹¹æœï¼Œ1è¡¨ç¤ºæ©™å­
y_train = np.array([0, 0, 0, 1, 1, 1])

# åˆ›å»ºæµ‹è¯•æ•°æ®
X_test = np.array([
    [170, 0],  # åº”è¯¥æ˜¯è‹¹æœ
    [270, 1],  # åº”è¯¥æ˜¯æ©™å­
])

# åˆ›å»ºå¹¶è®­ç»ƒKNNåˆ†ç±»å™¨
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# é¢„æµ‹
predictions = knn.predict(X_test)

# æ‰“å°ç»“æœ
for i, pred in enumerate(predictions):
    fruit = "è‹¹æœ" if pred == 0 else "æ©™å­"
    print(f"æµ‹è¯•æ ·æœ¬ {i+1} è¢«åˆ†ç±»ä¸º: {fruit}")
```

```
# åˆ›å»ºæ›´å¤æ‚çš„è®­ç»ƒæ•°æ®
# ç‰¹å¾ï¼šå¹´é¾„ã€æ”¶å…¥ã€ä¿¡ç”¨åˆ†æ•°
X_train = np.array([
    [25, 3000, 700],  # å¹´è½»äººï¼Œæ”¶å…¥ä¸€èˆ¬ï¼Œä¿¡ç”¨è‰¯å¥½
    [35, 8000, 750],  # ä¸­å¹´äººï¼Œæ”¶å…¥é«˜ï¼Œä¿¡ç”¨è‰¯å¥½
    [45, 12000, 800], # ä¸­å¹´äººï¼Œæ”¶å…¥å¾ˆé«˜ï¼Œä¿¡ç”¨ä¼˜ç§€
    [20, 2000, 600],  # å¹´è½»äººï¼Œæ”¶å…¥ä½ï¼Œä¿¡ç”¨ä¸€èˆ¬
    [30, 5000, 650],  # å¹´è½»äººï¼Œæ”¶å…¥ä¸­ç­‰ï¼Œä¿¡ç”¨ä¸€èˆ¬
    [40, 10000, 780], # ä¸­å¹´äººï¼Œæ”¶å…¥é«˜ï¼Œä¿¡ç”¨è‰¯å¥½
])

# æ ‡ç­¾ï¼š0è¡¨ç¤ºä¸æ‰¹å‡†ï¼Œ1è¡¨ç¤ºæ‰¹å‡†
y_train = np.array([0, 1, 1, 0, 0, 1])

# åˆ›å»ºå†³ç­–æ ‘åˆ†ç±»å™¨
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

# åˆ›å»ºæµ‹è¯•æ•°æ®
X_test = np.array([
    [28, 4000, 680],  # å¹´è½»äººï¼Œæ”¶å…¥ä¸­ç­‰ï¼Œä¿¡ç”¨ä¸€èˆ¬
    [38, 9000, 760],  # ä¸­å¹´äººï¼Œæ”¶å…¥é«˜ï¼Œä¿¡ç”¨è‰¯å¥½
])

# é¢„æµ‹
predictions = tree.predict(X_test)

# æ‰“å°ç»“æœ
for i, pred in enumerate(predictions):
    decision = "æ‰¹å‡†" if pred == 1 else "ä¸æ‰¹å‡†"
    print(f"æµ‹è¯•æ ·æœ¬ {i+1} çš„å†³å®šæ˜¯: {decision}")
```
