### å°æµ‹éªŒ

   - å¸®åŠ©æˆ‘ä»¬ç†è§£æ–‡å­—
   - è®©è®¡ç®—æœºå­¦ä¼š"è¯»"å’Œ"å†™"
   - æé«˜å·¥ä½œæ•ˆç‡

3. ä½ èƒ½æƒ³åˆ°å“ªäº›è‡ªç„¶è¯­è¨€å¤„ç†çš„ä¾‹å­ï¼Ÿ
   - æ¯”å¦‚ï¼šæœºå™¨ç¿»è¯‘ã€è¯­éŸ³åŠ©æ‰‹ã€æƒ…æ„Ÿåˆ†æ...

```
# å¯¼å…¥å¿…è¦çš„åº“
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import jieba
import re

# è®¾ç½®ä¸­æ–‡æ˜¾ç¤º
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

print("å·¥å…·å‡†å¤‡å®Œæ¯•ï¼è®©æˆ‘ä»¬å¼€å§‹å§ï¼")
```

## 2. æ–‡æœ¬åˆ†ç±»ï¼šè®©è®¡ç®—æœºå­¦ä¼šåˆ†ç±»æ–‡ç«  ğŸ“š

### ç”Ÿæ´»ä¸­çš„ä¾‹å­
- åƒåœ¾é‚®ä»¶è¿‡æ»¤
- æ–°é—»åˆ†ç±»
- è¯„è®ºæƒ…æ„Ÿåˆ†æ

è®©æˆ‘ä»¬ç”¨Pythonå®ç°ä¸€ä¸ªç®€å•çš„æ–‡æœ¬åˆ†ç±»å™¨ï¼

```
# å‡†å¤‡ä¸€äº›ç¤ºä¾‹æ–‡æœ¬
texts = [
    "è¿™éƒ¨ç”µå½±å¾ˆæ£’ï¼Œæˆ‘å¾ˆå–œæ¬¢ï¼",
    "è¿™ä¸ªäº§å“è´¨é‡å¾ˆå·®ï¼Œä¸æ¨èè´­ä¹°ã€‚",
    "ä»Šå¤©å¤©æ°”çœŸå¥½ï¼Œå¿ƒæƒ…æ„‰å¿«ã€‚",
    "æœåŠ¡æ€åº¦å¾ˆå·®ï¼Œè®©äººå¾ˆç”Ÿæ°”ã€‚",
    "è¿™ä¸ªé¤å…çš„èœå¾ˆå¥½åƒï¼Œä¸‹æ¬¡è¿˜æ¥ã€‚",
    "è¿™ä¸ªäº§å“æ€§ä»·æ¯”å¾ˆé«˜ï¼Œå€¼å¾—è´­ä¹°ã€‚",
    "è¿™éƒ¨ç”µå½±å¾ˆæ— èŠï¼Œæµªè´¹æ—¶é—´ã€‚",
    "è¿™ä¸ªæ™¯ç‚¹é£æ™¯ä¼˜ç¾ï¼Œå€¼å¾—ä¸€æ¸¸ã€‚"
]

# å‡†å¤‡æ ‡ç­¾ï¼ˆ1è¡¨ç¤ºæ­£é¢ï¼Œ0è¡¨ç¤ºè´Ÿé¢ï¼‰
labels = [1, 0, 1, 0, 1, 1, 0, 1]

# ä½¿ç”¨jiebaåˆ†è¯
def preprocess_text(text):
    # å»é™¤æ ‡ç‚¹ç¬¦å·
    text = re.sub(r'[^\w\s]', '', text)
    # åˆ†è¯
    words = jieba.cut(text)
    return ' '.join(words)

# é¢„å¤„ç†æ–‡æœ¬
processed_texts = [preprocess_text(text) for text in texts]

# åˆ›å»ºè¯è¢‹æ¨¡å‹
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(processed_texts)

# åˆ’åˆ†è®­ç»ƒé›†å’Œæµ‹è¯•é›†
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# åˆ›å»ºå¹¶è®­ç»ƒæœ´ç´ è´å¶æ–¯åˆ†ç±»å™¨
nb = MultinomialNB()
nb.fit(X_train, y_train)

# é¢„æµ‹
predictions = nb.predict(X_test)

# è®¡ç®—å‡†ç¡®ç‡
accuracy = accuracy_score(y_test, predictions)
print(f'å‡†ç¡®ç‡: {accuracy:.2f}')

# æ˜¾ç¤ºæ··æ·†çŸ©é˜µ
cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('æ··æ·†çŸ©é˜µ')
plt.xlabel('é¢„æµ‹æ ‡ç­¾')
plt.ylabel('çœŸå®æ ‡ç­¾')
plt.show()
```

## 4. æƒ…æ„Ÿåˆ†æï¼šè®©è®¡ç®—æœºç†è§£æƒ…æ„Ÿ ğŸ˜Š

### ç”Ÿæ´»ä¸­çš„ä¾‹å­
- è¯„è®ºæƒ…æ„Ÿåˆ†æ
- ç¤¾äº¤åª’ä½“æƒ…ç»ªåˆ†æ

è®©æˆ‘ä»¬å­¦ä¹ å¦‚ä½•åˆ†ææ–‡æœ¬æƒ…æ„Ÿï¼

```
# æƒ…æ„Ÿåˆ†æ
def analyze_sentiment(text):
    # é¢„å¤„ç†æ–‡æœ¬
    processed_text = preprocess_text(text)
    # è½¬æ¢ä¸ºç‰¹å¾å‘é‡
    features = vectorizer.transform([processed_text])
    # é¢„æµ‹æƒ…æ„Ÿ
    prediction = nb.predict(features)[0]
    return "æ­£é¢" if prediction == 1 else "è´Ÿé¢"

# æµ‹è¯•ä¸€äº›æ–°æ–‡æœ¬
test_texts = [
    "è¿™ä¸ªäº§å“å¾ˆå¥½ç”¨ï¼Œæ¨èè´­ä¹°ï¼",
    "æœåŠ¡æ€åº¦å¾ˆå·®ï¼Œä¸æ¨èã€‚",
    "è¿™ä¸ªæ™¯ç‚¹é£æ™¯ä¼˜ç¾ï¼Œå€¼å¾—ä¸€æ¸¸ã€‚"
]

for text in test_texts:
    sentiment = analyze_sentiment(text)
    print(f'æ–‡æœ¬: {text}')
    print(f'æƒ…æ„Ÿ: {sentiment}\n')
```
